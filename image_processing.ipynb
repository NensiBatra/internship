```python
cv2.waitKey(0)
height, width = img.shape[0:2]
img.shape[0:2]
print(img.shape[0:2])
rotation_matrix = cv2.getRotationMatrix2D((width/2, height/2), 90, .5)
rotated_image = cv2.warpAffine(img, rotation_matrix, (width, height))

cv2_imshow(rotated_image)
cv2.waitKey(0)
```

```python

start_row = int(height*.15)
end_row = int(height*.7)
start_col = int(width*.15)
end_col = int(width*.85)
cropped_image = img[start_row:end_row, start_col:end_col]

cv2_imshow(img)
cv2_imshow(cropped_image)
cv2.waitKey(0)
```

```python
from google.colab.patches import cv2_imshow
import cv2

img = cv2.imread("/content/drive/MyDrive/internship/emoji.jpeg")
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
moment = cv2.moments(gray_img)
X = int(moment["m10"] / moment["m00"])
Y = int(moment["m01"] / moment["m00"])
cv2.circle(img, (X, Y), 15, (205, 114, 101), 1)
cv2_imshow(img)
cv2.waitKey(0)
```
